```python
import pandas as pd
```

```python
wine=pd.read_csv('Data/winemag-data-130k-v2.csv')
```

```python
wine.info()
```

```python
chunk=wine[wine['price']<18]
```

```python
chunk.nlargest(5,'points')
```

```python
chunk.nsmallest(5,'points')
```

```python
#find the most generous and least generous taster
wine.groupby('taster_name')['points'].mean().sort_values(ascending=False)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
wine.groupby('taster_name')['points'].mean().sort_values(ascending=False).plot(kind='bar',figsize=(12,5),title='Tasters Scores',color='blue')
plt.ylim(80,93)
```
